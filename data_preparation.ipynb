# Data preparation

```python
import pandas as pd
import pickle as rick
from sklearn.preprocessing import StandardScaler
```

```python
# Set labels
labels = [
    'ID',
    'Diagnosis',
    'c1-radius',
    'c1-texture',
    'c1-perimeter',
    'c1-area',
    'c1-smoothness',
    'c1-compactness',
    'c1-concavity',
    'c1-concave points',
    'c1-symmetry',
    'c1-fractal dimension',
    'c2-radius',
    'c2-texture',
    'c2-perimeter',
    'c2-area',
    'c2-smoothness',
    'c2-compactness',
    'c2-concavity',
    'c2-concave points',
    'c2-symmetry',
    'c2-fractal dimension',
    'c3-radius',
    'c3-texture',
    'c3-perimeter',
    'c3-area',
    'c3-smoothness',
    'c3-compactness',
    'c3-concavity',
    'c3-concave points',
    'c3-symmetry',
    'c3-fractal dimension',
]
```

```python
# Import with pandas
df = pd.read_csv('data/wdbc.data', names=labels, encoding= 'unicode_escape')
```

```python
# Set target 'y'
y = df['Diagnosis'].tolist()

# Replace 'M' with 0 and 'B' with 1
y = [0 if i == 'M' else 1 for i in y] 

# And show top-25
y[:25]
```

```python
# Set data 'X' (by removing target column)
X = df.drop(columns=['ID', 'Diagnosis'])

# And show top-10
X[:10]
```

```python
# Use Z-scale transformation
scaler = StandardScaler()

scaler.fit(X)
```

```python
# Show transformed data
X = scaler.transform(X)

print(X)
```

```python
# Save 'X' as pickle object
with open('X.pickle', 'wb') as f:
    rick.dump(X, f)

# Save 'y' as pickle object
with open('y.pickle', 'wb') as f:
    rick.dump(y, f)
```

### Dimension reduction

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
pca.fit(X)
PCA(n_components=2)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
# Create PCA dot plot
X_reduced = PCA(n_components=2).fit_transform(X)
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)

ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    edgecolor="k",
    s=40,
)

ax.set_title("First two PCA directions")
ax.set_xlabel("1st eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.yaxis.set_ticklabels([])
#plt.show()
plt.savefig('pca.png')
```
